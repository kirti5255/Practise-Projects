# Data Cleaning

```python
#removing outliers
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
#dropping outliers
#drop outliers
df_new=df[(z<3).all(axis=1)]
```

```python
df_new.shape
```

```python
df.shape
```

```python
#some of the outliers have been removed.
```

```python
df_new
```

```python
#to remove skewness
from scipy.stats import boxcox 
df_new['Family']=boxcox(df_new['Family'], 0) #log transform method
```

```python
#taking only input data
x=df_new.iloc[:,1:]
x.head()
```

```python
x.shape
```

```python
#taking only output data
y=df_new.iloc[:,0:1]
y.head()
```

```python
y.shape
```

# Perform Regression

```python
#70% training 30% for test- splitting
#random state is the seed value 42, 45 is defalut
#outliers are removed
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=.30, random_state=42)
```

```python
x_train.shape #training data: 107 rows to go for trainning
```

```python
x_test.shape #test data: 46 rows go for testing
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train, y_train)  #training it
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_train, y_train)
```

```python
#machine working 78.14% correct
```

```python
#predict the value
pred=lm.predict(x_test)
```

```python
print("predicted result price: ", pred)
print("actual price: ", y_test)
```

```python
print('error:')
print("mean absolute error:", mean_absolute_error(y_test,pred))
print("mean squared error:", mean_squared_error(y_test,pred))
print("root mean square error:", np.sqrt(mean_squared_error(y_test,pred)))
```

```python
#r2 score - how much is change coming in dependednt variable when independent variable is changing

from sklearn.metrics import r2_score
print(r2_score(y_test, pred))
```
